```python
import pandas as pd
df=pd.read_excel("spreadsheets/students.xlsx")
#iii> Calculate percentage of each student and display it.
df["Percentage"]=(df["Maths"]+df["Physics"]+df["Chemistry"]+df["C.S."]+df["English"])/5
df=df[["Name","Percentage"]]
df.head(50)
```
